```python
import pandas as pd
import numpy as np
```

```python
files = ['rollingsales_bronx.xlsx',
         'rollingsales_brooklyn.xlsx',
         'rollingsales_manhattan.xlsx',
         'rollingsales_queens.xlsx',
         'rollingsales_statenisland.xlsx']
```

```python
df1 = pd.DataFrame()

for file in files:
    df_file = pd.read_excel(file, skiprows=4)
    df1 = df1.append(df_file, ignore_index=True)
```

```python
df1.head()
```

```python
df1['BOROUGH'].unique()
```

```python
b_map = {2: 'Bronx', 3: 'Brooklyn', 1:'Manhattan', 4:'Queens', 5:'Staten Island'}
```

```python
df1=df1.replace({'BOROUGH':b_map})
df1
```

```python
df1.describe()
```

```python
df1.dtypes
```

```python
df1.columns
```

```python
df = df1[['BOROUGH', 'NEIGHBORHOOD', 'BUILDING CLASS CATEGORY', 'BLOCK', 'LOT',
      'ADDRESS', 'ZIP CODE','LAND SQUARE FEET', 'GROSS SQUARE FEET', 'YEAR BUILT',
       'SALE PRICE', 'SALE DATE']]
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.isna().sum()
```

```python
df[df['ZIP CODE'].isnull()]
```

```python
df['ZIP CODE'] = df['ZIP CODE'].fillna(method='ffill')
```

```python
df['ZIP CODE'] = df['ZIP CODE'].astype('int')
```

```python
df = df[df['SALE PRICE'] != 0]
```

```python
df.head()
```

```python
df.to_csv('nyc_rolling_sales.csv', index=False)
```
